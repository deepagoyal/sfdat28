```python
def parse_beer(beer_dict):
    # given dict of beer from API
    # return row of dataframe
    to_return = {}
    for col in ['name', 'abv', 'ibu', 'description']:
        to_return[col] = beer_dict.get(col)
    to_return['glass_type'] = beer_dict.get('glass', {}).get('name')
    to_return['style'] = beer_dict.get('style', {}).get('name')
    to_return['category'] = beer_dict.get('style', {}).get('category', {}).get('name')
    to_return['label'] = beer_dict.get('labels', {}).get('large')
    return to_return
```
